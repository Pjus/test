```python
plt.style.use('ggplot')
plt.ion()
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.set_xlim(0,31)

xs = deque(maxlen=31)
ys = deque(maxlen=31)
identifier = "AAPL"
def animate(i):
    data = pd.read_csv('./data.csv')
    num = len(data)
#     print("len : {}".format(num))
    
    x_deque = deque(maxlen=25)
    y_deque = deque(maxlen=25)

    x = data['Date'].values
    y = data['Close'].values

    for i in x:
        x_deque.append(i)
    for i in y:
        y_deque.append(i)
    
    plt.cla()
    plt.ylabel('Price')
    plt.title('Stock Code: {}'.format(identifier))
    plt.plot(x_deque, y_deque, alpha=0.8, label="Close")
    
    plt.xticks(rotation=45)

    
    plt.legend(loc='upper left')
    plt.tight_layout()

```

```python
ani = animation.FuncAnimation(plt.gcf(), animate, interval=500)
ani.save('stock2.gif', writer='imagemagick', fps=30)
plt.show()
```
